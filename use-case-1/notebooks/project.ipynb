## Inferenz

```python
import config
import torch
import torchvision.transforms.functional as TF
import albumentations as A
from albumentations.pytorch import ToTensorV2
from PIL import Image
import os
import numpy as np

import matplotlib.pyplot as plt
import numpy as np

def load_model_and_predict(image_folder, image_filename, checkpoint_path, device):
    # Laden des Modells
    model = mu.create_model()
    model.to(device)

    # Laden des Modell-Checkpoints
    model.load_state_dict(torch.load(checkpoint_path, map_location=device))
    #print(model)
    model.eval()

    # Bild laden
    img_path = os.path.join(image_folder, image_filename)
    image = np.array(Image.open(img_path).convert('RGB'))  # Konvertieren zu numpy array

    # Albumentations Transformationen definieren
    preprocessing = A.Compose([
        A.Resize(height=config.IMAGE_HEIGHT, width=config.IMAGE_WIDTH),
        A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225], max_pixel_value=255.0,),
        ToTensorV2(),
    ])

    # Bild transformieren
    transformed = preprocessing(image=image)
    image = transformed['image'].unsqueeze(0).to(device)  # Unsqueeze fgt eine Batch-Dimension hinzu

    # Vorhersage
    with torch.no_grad():
        output = model(image).cpu()
        mask = torch.sigmoid(output) > config.THRESHOLD
        mask = TF.resize(
                mask, size=(480, 608), interpolation=TF.InterpolationMode.NEAREST
            )
        #print(mask[0].squeeze().numpy().shape)
    return mask[0].squeeze().numpy()



def show_prediction(image_folder, image_filename, mask):
    img_path = os.path.join(image_folder, image_filename)
    image = np.array(Image.open(img_path).convert('RGB'))  
    label_path = img_path.replace('imgs', 'masks').replace('.jpg', '.gif')
    label = np.array(Image.open(label_path))
    label[label == 1] = 255

    plt.figure(figsize=(20, 10))
    # Zeige das Originalbild
    plt.subplot(1, 3, 1) 
    plt.title('Originalbild')
    plt.imshow(image)
    plt.axis('off')

    # Zeige das Label
    plt.subplot(1, 3, 2)
    plt.title('Grundwahrheit')
    plt.imshow(label, cmap='gray')
    plt.axis('off')

    # Zeige die Vorhersagemaske
    plt.subplot(1, 3, 3)
    plt.title('Vorhersagemaske')
    plt.imshow(mask, cmap='gray')
    plt.axis('off')

    plt.show()



# Beispiel
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
checkpoint_path = './../models/trained_models/best_model.pth'
image_folder = './../data/raw/test/imgs/'
image_filename = '1474327855_20023.jpg'
prediction = load_model_and_predict(image_folder, image_filename, checkpoint_path, device)
show_prediction(image_folder, image_filename, prediction)
```
